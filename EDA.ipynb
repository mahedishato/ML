```python
df_country = pd.read_excel('Zom/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on='Country Code', how= 'left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df['Country'].value_counts().index
country_names
```

```python
for_pie=final_df['Country'].value_counts().values
for_pie
```

```python
plt.pie(for_pie[:3], labels=country_names[:3])
```

```python
plt.pie(final_df['Country'].value_counts()[:2], labels=final_df['Country'].value_counts().index[:2] , autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
nn=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
nn
```

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize']= (16,6)
sns.barplot(x=nn['Aggregate rating'], y= nn['Rating_count'])
```

```python
matplotlib.rcParams['figure.figsize']= (16,6)
sns.barplot(x=nn['Aggregate rating'], y= nn['Rating_count'], hue=nn['Rating color'], palette=['b','Red','Orange','Yellow', 'Green','Green'])
```
